## 4. Input and Output Operations

Programs need to interact with users!

### 4.1 Output with print()
We've already seen print(). Let's explore more features:

## 5. Basic Operators

Operators let us perform operations on data

### 5.1 Arithmetic Operators
For mathematical calculations

```python
# Arithmetic operators
a = 10
b = 3

print("Addition:", a + b)        # 13
print("Subtraction:", a - b)     # 7
print("Multiplication:", a * b)  # 30
print("Division:", a / b)        # 3.333... (always returns float)
print("Floor Division:", a // b) # 3 (rounds down to integer)
print("Modulus (remainder):", a % b)  # 1
print("Exponentiation:", a ** b) # 1000 (10 to the power of 3)
```

### 5.2 Comparison Operators
Compare values and return True or False

```python
# Comparison operators
x = 5
y = 10

print("Equal to:", x == y)           # False
print("Not equal to:", x != y)       # True
print("Greater than:", x > y)        # False
print("Less than:", x < y)           # True
print("Greater or equal:", x >= y)   # False
print("Less or equal:", x <= y)      # True
```

### 5.3 Logical Operators
Combine multiple conditions

```python
# Logical operators
is_sunny = True
is_warm = False

print("AND:", is_sunny and is_warm)  # False (both must be True)
print("OR:", is_sunny or is_warm)    # True (at least one is True)
print("NOT:", not is_sunny)          # False (reverses the value)
```

## 6. Control Structures: Conditional Statements

Make decisions in your code!

### 6.1 if Statement
Execute code only if a condition is True

```python
# Simple if statement
age = 18

if age >= 18:
    print("You are an adult!")  # This line is indented - it's inside the if block
    print("You can vote!")

print("This line always runs")  # Not indented - outside the if block
```

### 6.2 if-else Statement
Do one thing if True, another if False

```python
# if-else statement
temperature = 25

if temperature > 30:
    print("It's hot outside!")
else:
    print("The weather is pleasant.")
```

### 6.3 if-elif-else Statement
Check multiple conditions

```python
# if-elif-else statement
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your grade is: {grade}")
```

### 6.4 Nested if Statements
if statements inside other if statements

```python
# Nested if statements
age = 20
has_license = True

if age >= 18:
    print("You are old enough to drive.")
    if has_license:
        print("You can drive!")
    else:
        print("But you need a license first.")
else:
    print("You are too young to drive.")
```

## 7. Control Structures: Loops

Repeat code multiple times!

### 7.1 for Loop
Loop through a sequence (like a range of numbers)

```python
# Basic for loop with range()
# range(5) gives us: 0, 1, 2, 3, 4

print("Counting from 0 to 4:")
for i in range(5):
    print(i)
```

```python
# range() with start and end
# range(1, 6) gives us: 1, 2, 3, 4, 5

print("Counting from 1 to 5:")
for i in range(1, 6):
    print(i)
```

```python
# range() with step
# range(0, 11, 2) gives us: 0, 2, 4, 6, 8, 10

print("Even numbers from 0 to 10:")
for i in range(0, 11, 2):
    print(i)
```

```python
# Looping through a string
name = "Python"

print("Each letter in Python:")
for letter in name:
    print(letter)
```

### 7.2 while Loop
Keep looping while a condition is True

```python
# Basic while loop
count = 1

print("Counting with while loop:")
while count <= 5:
    print(count)
    count = count + 1  # or count += 1

print("Loop finished!")
```

```python
# while loop with user input
# This creates a simple menu system

choice = ""

while choice != "quit":
    choice = input("Enter 'hello' to greet or 'quit' to exit: ")
    
    if choice == "hello":
        print("Hello there!")
    elif choice == "quit":
        print("Goodbye!")
    else:
        print("I don't understand that command.")
```

### 7.3 break and continue
Control loop execution

```python
# break - exits the loop immediately
print("Using break:")
for i in range(10):
    if i == 5:
        break  # Stop the loop when i is 5
    print(i)

print("Loop stopped at 5")
```

```python
# continue - skips the rest of the current iteration
print("Using continue (skip odd numbers):")
for i in range(10):
    if i % 2 != 0:  # If i is odd
        continue  # Skip to next iteration
    print(i)  # Only prints even numbers
```

## 8. Functions

Functions are reusable blocks of code that perform specific tasks

### 8.1 Defining and Calling Functions
Create your own functions!

```python
# Simple function with no parameters
def greet():
    """This function prints a greeting"""
    print("Hello! Welcome to Python functions!")

# Call the function
greet()
greet()  # We can call it multiple times
```

### 8.2 Functions with Parameters
Pass data to functions

```python
# Function with one parameter
def greet_person(name):
    """Greet a person by name"""
    print(f"Hello, {name}!")

# Call with different arguments
greet_person("Alice")
greet_person("Bob")
greet_person("Charlie")
```

```python
# Function with multiple parameters
def add_numbers(a, b):
    """Add two numbers and print the result"""
    result = a + b
    print(f"{a} + {b} = {result}")

add_numbers(5, 3)
add_numbers(10, 20)
```

### 8.3 Return Values
Functions can send data back to the caller

```python
# Function that returns a value
def multiply(a, b):
    """Multiply two numbers and return the result"""
    return a * b

# Store the returned value
result = multiply(4, 5)
print(f"Result: {result}")

# Use the return value directly
print(f"6 times 7 is {multiply(6, 7)}")
```

```python
# Function with multiple return values
def calculate(a, b):
    """Perform multiple calculations"""
    addition = a + b
    subtraction = a - b
    multiplication = a * b
    return addition, subtraction, multiplication

# Unpack the returned values
add, sub, mul = calculate(10, 5)
print(f"Addition: {add}")
print(f"Subtraction: {sub}")
print(f"Multiplication: {mul}")
```

### 8.4 Default Parameters
Give parameters default values

```python
# Function with default parameter
def greet_with_title(name, title="Friend"):
    """Greet someone with an optional title"""
    print(f"Hello, {title} {name}!")

# Call with and without the optional parameter
greet_with_title("Alice")              # Uses default title
greet_with_title("Bob", "Dr.")         # Uses provided title
greet_with_title("Charlie", "Professor")
```

### 8.5 Variable Scope
Where variables can be accessed

```python
# Local vs Global variables

# Global variable (defined outside functions)
global_var = "I'm global!"

def my_function():
    # Local variable (only exists inside this function)
    local_var = "I'm local!"
    print(local_var)
    print(global_var)  # Can access global variables

my_function()
print(global_var)  # Works
# print(local_var)  # This would cause an error!
```

```python
# Modifying global variables inside functions
counter = 0

def increment():
    global counter  # Tell Python we want to modify the global variable
    counter += 1
    print(f"Counter is now: {counter}")

increment()
increment()
increment()
print(f"Final counter value: {counter}")
```

## 9. PROJECT 1: "Guess the Number" Game

Let's put everything together in a fun project!

**Game Rules:**
1. Computer picks a random number between 1 and 100
2. Player tries to guess the number
3. Computer gives hints (too high or too low)
4. Game continues until player guesses correctly
5. Display the number of attempts taken

### Step 1: Import the random module
We need this to generate a random number

```python
# Import the random module
import random

# Test it out
print("Random number between 1 and 10:", random.randint(1, 10))
print("Another random number:", random.randint(1, 10))
```

### Step 2: Generate the secret number
Computer picks a random number

```python
# Generate a random number between 1 and 100
secret_number = random.randint(1, 100)

# For testing, let's see what it is (remove this in final game!)
print(f"Secret number (for testing): {secret_number}")
```

### Step 3: Initialize game variables
Set up variables to track the game state

```python
# Variables to track the game
attempts = 0          # Count how many guesses
guessed = False       # Has the player won?

print("Variables initialized!")
print(f"Attempts: {attempts}")
print(f"Guessed: {guessed}")
```

### Step 4: Create the game loop
Keep asking for guesses until player wins

```python
# Simple version - just one guess for testing
guess = int(input("Enter your guess (1-100): "))
attempts = 1

if guess == secret_number:
    print("Correct! You guessed it!")
elif guess < secret_number:
    print("Too low! Try again.")
else:
    print("Too high! Try again.")
```

### Step 5: Complete Game with Loop
Full game that keeps going until you win!

```python
# Complete game loop
import random

secret_number = random.randint(1, 100)
attempts = 0
guessed = False

print("Welcome to Guess the Number!")
print("I'm thinking of a number between 1 and 100.")
print()

while not guessed:
    # Get player's guess
    guess = int(input("Enter your guess: "))
    attempts += 1
    
    # Check the guess
    if guess == secret_number:
        guessed = True
        print(f"\n🎉 Congratulations! You guessed it!")
        print(f"The number was {secret_number}")
        print(f"It took you {attempts} attempts.")
    elif guess < secret_number:
        print("📈 Too low! Try a higher number.\n")
    else:
        print("📉 Too high! Try a lower number.\n")
```
